## drop columns that will not be used in the fitting model

```python
data_df=df.drop(['rowid','kepid','kepoi_name','kepler_name'], axis=1)
data_df
```

```python
data_df=data_df.drop(["koi_tce_delivname",'koi_teq_err1','koi_teq_err2'], axis=1)
```

```python
data_df
```

```python
from numpy import asarray
from sklearn.preprocessing import MinMaxScaler
```

## Defining domain of the function (X) or features. All the df columns except the column to be predicted
## Defining range of the function (y) or target. One of the columns that was dropped in the step above

```python
X=data_df.drop(['koi_disposition','koi_pdisposition'], axis=1)
# y=data_df['koi_disposition'].values.reshape(-1,1)
target=data_df['koi_disposition'].dropna()
X=X.dropna()
```

```python
9564-7803
```

```python
target=target[:7803]

print(X.shape,target.shape)
```

```python
y=target
print(X.shape,y.shape)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

## Create a MinMax model and fit it to the training data

```python
from sklearn.preprocessing import StandardScaler

X_scaler = StandardScaler().fit(X)
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
from sklearn.svm import SVC 

model = SVC(kernel='rbf',gamma=.5 )
model.fit(X_train_scaled, y_train)
predictions = model.predict(X_test_scaled)
```

```python
model
```

```python
print(f"Training Data Score for rbf model: {model.score(X_train_scaled, y_train)}")
print(f"Testing Data Score for rbf model: {model.score(X_test_scaled, y_test)}")
```

```python
predictions = model.predict(X_test_scaled)
print(f"First 10 Predictions:   {predictions[:10]}")
print(f"First 10 Actual labels: {y_test[:10].tolist()}") 
```

```python
first_pass = pd.DataFrame({"Prediction": predictions, "Actual": y_test}).reset_index(drop=True)
first_pass
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, predictions, 
                            target_names = ["Confirmed", "False Positive", "Candidate"]))
```

```python
from sklearn.model_selection import GridSearchCV
param_grid = {'C': [1],
              'gamma':[0.01,0.1,0.99] }
grid = GridSearchCV(model, param_grid, verbose=3)
```

# Train the model with GridSearch

```python
grid.fit(X_train, y_train)
```
