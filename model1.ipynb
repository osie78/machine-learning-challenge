```python
df=pd.read_csv('data/cumulative.csv')
```

```python
# drop columns that will not be used in the fitting model
data_df=df.drop(['rowid','kepid','kepoi_name','kepler_name'], axis=1)


data_df
```

```python
data_df.dtypes
```

```python
data_df=data_df.drop(["koi_tce_delivname",'koi_teq_err1','koi_teq_err2'], axis=1)
```

```python
data_df
```

```python
from numpy import asarray
from sklearn.preprocessing import MinMaxScaler
```

```python
# Defining domain of the function (X) or features. All the df columns except the column to be predicted
# Defining range of the function (y) or target. One of the columns that was dropped in the step above
```

```python
X=data_df.drop(['koi_disposition','koi_pdisposition'], axis=1)
# y=data_df['koi_disposition'].values.reshape(-1,1)
target=data_df['koi_disposition'].dropna()
X=X.dropna()
```

```python
9564-7803
```

```python
target=target[:7803]

print(X.shape,target.shape)
```

```python
y=target
print(X.shape,y.shape)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(max_iter=10000)
classifier
```

```python
from sklearn.preprocessing import StandardScaler

# Create a MinMax model and fit it to the training data

X_scaler = MinMaxScaler().fit(X)
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
classifier.fit(X_train_scaled, y_train)
```

```python
print(f"Training Data Score: {classifier.score(X_train_scaled, y_train)}")
print(f"Testing Data Score: {classifier.score(X_test_scaled, y_test)}")
```

```python
predictions = classifier.predict(X_test_scaled)
print(f"First 10 Predictions:   {predictions[:10]}")
print(f"First 10 Actual labels: {y_test[:10].tolist()}") 
```

```python
first_pass = pd.DataFrame({"Prediction": predictions, "Actual": y_test}).reset_index(drop=True)
first_pass
```
