## Defining range of the function (y) or target. One of the columns that was dropped in the step above

```python
X=data_df.drop(['koi_disposition','koi_pdisposition'], axis=1)

target=data_df['koi_disposition'].dropna()
X=X.dropna()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

## Create a MinMax model and fit it to the training data

```python
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.utils import to_categorical

X_scaler = MinMaxScaler().fit(X)
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python

from sklearn.preprocessing import LabelEncoder # One-hot encoding
label_encoder = LabelEncoder()
label_encoder.fit(y_train)

encoded_y_train = label_encoder.transform(y_train)
encoded_y_test = label_encoder.transform(y_test)

y_train_categorical = to_categorical(encoded_y_train)
y_test_categorical = to_categorical(encoded_y_test)

```

```python
# first, create a normal neural network with 2 inputs, 6 hidden nodes, and 2 outputs
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

model = Sequential()
model.add(Dense(units=192, activation='relu', input_dim=32))
model.add(Dense(units=192, activation='relu'))
model.add(Dense(units=192, activation='relu'))

model.add(Dense(units=3, activation='softmax'))
```

```python
model.summary()
```

### Compile the model

```python
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])
```

### Fit the model to the training data

```python

model.fit(
    X_train_scaled,
    y_train_categorical,
    epochs=100,
    shuffle=True,
    verbose=2
)
```
